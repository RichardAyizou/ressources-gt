```python
import pandas as pd
import numpy as np
```

```python
text_total_district = pd.read_table("../text_total_district.txt", sep=",", header=None)

col_text_total_district = {
    0: "district_depart",
    1: "district_destination",
    2: "nombre_text"
}
text_total_district.rename(columns=col_text_total_district, inplace=True)

df_total_text = pd.DataFrame({"id_district":range(0,124)})

df = text_total_district
# total de message 
df_total_text["nb_msg"]= df[["district_depart", "nombre_text"]].groupby("district_depart").sum()
# nombre moyen de message qui sort de chaque district par heure
df_total_text["nb_moyen_msg"]= df[["district_depart", "nombre_text"]].groupby("district_depart").mean()
# standard deviation of message send from each district by hour
#df_test["std_msg_{}h".format(h)]= np.std(df["nombre_text"])
# nombre minimum de message qui sort de chaque district par heure
df_total_text["nb_min_msg"]= df[["district_depart", "nombre_text"]].groupby("district_depart").min()
# nombre maximum de message qui sort de chaque district par heure
df_total_text["nb_max_msg"]= df[["district_depart", "nombre_text"]].groupby("district_depart").max()
# transformation logarithmique
df_total_text["log_nb_msg"]= np.log(df[["district_depart", "nombre_text"]].groupby("district_depart").sum())
df_total_text["log_nb_moyen_msg"]= np.log(df[["district_depart", "nombre_text"]].groupby("district_depart").mean())
# transformation racine carrée
df_total_text["sqrt_nb_msg"]= np.sqrt(df[["district_depart", "nombre_text"]].groupby("district_depart").sum())
df_total_text["sqrt_nb_moyen_msg"]= np.sqrt(df[["district_depart", "nombre_text"]].groupby("district_depart").mean())


df_total_text = df_total_text.iloc[1:,:]
#df_total_text.head()
```
