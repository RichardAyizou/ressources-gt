```python
import pandas as pd
import numpy as np
```

```python
text_weekday_district = pd.read_table("../text_weekday_district.txt", sep=",", header=None)

col_text_weekday_district = {
    0: "jour_semaine",
    1: "district_depart",
    2: "district_destination",
    3 : "nombre_text"
}
text_weekday_district.rename(columns=col_text_weekday_district, inplace=True)

df_weekday_text = pd.DataFrame({"id_district":range(0,124)})

for j in range(1,8):
    # filtrer par weekday
    df = text_weekday_district[text_weekday_district["jour_semaine"]==j]
    # total de message que chaque district envoie aux autres chaque heure
    df_weekday_text["nb_msg_j{}".format(j)]= df[["district_depart", "nombre_text"]].groupby("district_depart").sum()
    # nombre moyen de message qui sort de chaque district par heure
    df_weekday_text["nb_moyen_msg_j{}".format(j)]= df[["district_depart", "nombre_text"]].groupby("district_depart").mean()
    # standard deviation of message send from each district by hour
    #df_test["std_msg_{}h".format(h)]= np.std(df["nombre_text"])
    # nombre minimum de message qui sort de chaque district par heure
    df_weekday_text["nb_min_msg_j{}".format(j)]= df[["district_depart", "nombre_text"]].groupby("district_depart").min()
    # nombre maximum de message qui sort de chaque district par heure
    df_weekday_text["nb_max_msg_j{}".format(j)]= df[["district_depart", "nombre_text"]].groupby("district_depart").max()
    # transformation logarithmique
    df_weekday_text["log_nb_msg_j{}".format(j)]= np.log(df[["district_depart", "nombre_text"]].groupby("district_depart").sum())
    df_weekday_text["log_nb_moyen_msg_j{}".format(j)]= np.log(df[["district_depart", "nombre_text"]].groupby("district_depart").mean())
    # transformation racine carrée
    df_weekday_text["sqrt_nb_msg_j{}".format(j)]= np.sqrt(df[["district_depart", "nombre_text"]].groupby("district_depart").sum())
    df_weekday_text["sqrt_nb_moyen_msg_j{}".format(j)]= np.sqrt(df[["district_depart", "nombre_text"]].groupby("district_depart").mean())
    

df_weekday_text = df_weekday_text.iloc[1:,:]
#df_weekday_text.head()
```
