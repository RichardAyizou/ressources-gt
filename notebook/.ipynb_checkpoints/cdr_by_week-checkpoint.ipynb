```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
text_week_district = pd.read_table("../text_week_district.txt", sep=",", header=None)

col_text_week_district ={
    0: "semaine",
    1: "district_depart",
    2: "district_destination",
    3 : "nombre_text"
}
text_week_district.rename(columns=col_text_week_district, inplace=True)

df_week_text = pd.DataFrame({"id_district":range(0,124)})

for s in range(1,53):
    # filtrer par semaine
    df = text_week_district[text_week_district["semaine"]==s]
    # total de message que chaque district envoie aux autres chaque semaine
    df_week_text["nb_msg_s{}".format(s)]= df[["district_depart", "nombre_text"]].groupby("district_depart").sum()
    # nombre moyen de message qui sort de chaque district par semaine
    df_week_text["nb_moyen_msg_s{}".format(s)]= df[["district_depart", "nombre_text"]].groupby("district_depart").mean()
    # nombre minimum de message qui sort de chaque district par semaine
    df_week_text["nb_min_msg_s{}".format(s)]= df[["district_depart", "nombre_text"]].groupby("district_depart").min()
    # nombre maximum de message qui sort de chaque district par semaine
    df_week_text["nb_max_msg_s{}".format(s)]= df[["district_depart", "nombre_text"]].groupby("district_depart").max()
    # transformation logarithmique
    df_week_text["log_nb_msg_s{}".format(s)]= np.log(df[["district_depart", "nombre_text"]].groupby("district_depart").sum())
    df_week_text["log_nb_moyen_msg_s{}".format(s)]= np.log(df[["district_depart", "nombre_text"]].groupby("district_depart").mean())
    # transformation racine carrée
    df_week_text["sqrt_nb_msg_s{}".format(s)]= np.sqrt(df[["district_depart", "nombre_text"]].groupby("district_depart").sum())
    df_week_text["sqrt_nb_moyen_msg_s{}".format(s)]= np.sqrt(df[["district_depart", "nombre_text"]].groupby("district_depart").mean())
    

df_week_text = df_week_text.iloc[1:,:]
df_week_text.head()
```

```python
voice_week_district_duration = pd.read_table("../voice_week_district_duration.txt", sep=",", header=None)


col_voice_week_district_duration = {
    0: "semaine",
    1: "district_depart",
    2: "district_destination",
    3: "duree_voice_semaine"
}
voice_week_district_duration.rename(columns=col_voice_week_district_duration, inplace=True)


df_week_voice_duree = pd.DataFrame({"id_district":range(0,124)})

for s in range(1,53):
    # filtrer par semaine
    df = voice_week_district_duration[voice_week_district_duration["semaine"]==s]
    # total de message que chaque district envoie aux autres chaque semaine
    df_week_voice_duree["duree_voice_s{}".format(s)]= df[["district_depart", "duree_voice_semaine"]].groupby("district_depart").sum()
    # nombre moyen de message qui sort de chaque district par semaine
    df_week_voice_duree["duree_moyen_voice_s{}".format(s)]= df[["district_depart", "duree_voice_semaine"]].groupby("district_depart").mean()
    # nombre minimum de message qui sort de chaque district par semaine
    df_week_voice_duree["duree_min_voice_s{}".format(s)]= df[["district_depart", "duree_voice_semaine"]].groupby("district_depart").min()
    # nombre maximum de message qui sort de chaque district par semaine
    df_week_voice_duree["duree_max_voice_s{}".format(s)]= df[["district_depart", "duree_voice_semaine"]].groupby("district_depart").max()
    # transformation logarithmique
    df_week_voice_duree["log_duree_voice_s{}".format(s)]= np.log(df[["district_depart", "duree_voice_semaine"]].groupby("district_depart").sum())
    df_week_voice_duree["log_duree_moyen_voice_s{}".format(s)]= np.log(df[["district_depart", "duree_voice_semaine"]].groupby("district_depart").mean())
    # transformation racine carrée
    df_week_voice_duree["sqrt_duree_voice_s{}".format(s)]= np.sqrt(df[["district_depart", "duree_voice_semaine"]].groupby("district_depart").sum())
    df_week_voice_duree["sqrt_duree_moyen_voice_s{}".format(s)]= np.sqrt(df[["district_depart", "duree_voice_semaine"]].groupby("district_depart").mean())
    

df_week_voice_duree = df_week_voice_duree.iloc[1:,:]
df_week_voice_duree.head()
```

```python
df_week = pd.DataFrame({"id_district": range(1,124)})

for my_df in [df_week_text, df_week_voice, df_week_voice_duree]:
    df_week = pd.merge(df_week, my_df, on="id_district")

df_week.head()
```
